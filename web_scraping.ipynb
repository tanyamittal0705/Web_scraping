# web scraping

```python
import requests
page = requests.get("http://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168")
page
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
seven_day=soup.find(id="seven-day-forecast")
seven_day
```

```python
forecast_items=seven_day.find_all(class_="tombstone-container")
print(len(forecast_items))
print(forecast_items)
```

```python
a=seven_day.select(".tombstone-container .period-name")
a
```

```python
a[0].text
```

```python
b=[x.text for x in a]
print(b)
```

```python
print(len(forecast_items),"\n")
forecast_items[0]
```

```python
today=forecast_items[0]
print(today.prettify())
```

```python
period=today.find(class_="period-name").get_text()
```

```python
period
```

```python
short_desc = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
print(short_desc)
```

```python
temp = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
print(temp)
```

```python
descs=[d["title"] for d in seven_day.select(".tombstone-container img")]
print(descs)
```
